```python
import os
from pathlib import Path

def create_project_structure(project_name):
    """Create a well-organized directory structure for your research project"""
    
    # Create the main project folder
    project_path = Path(project_name)
    project_path.mkdir(exist_ok=True)
    
    # Create data and scripts directories at the project root
    data_path = project_path / "data"
    scripts_path = project_path / "scripts"
    
    data_path.mkdir(exist_ok=True)
    scripts_path.mkdir(exist_ok=True)
    
    # my6-folder system inside the data directory
    data_folders = [
        "raw_data"
    ]
    
    # Create each data subfolder
    for folder in data_folders:
        folder_path = data_path / folder
        folder_path.mkdir(exist_ok=True)
        
        # Add a README to explain each folder's purpose
        readme_path = folder_path / "README.txt"
        
        # Write helpful descriptions
        if "raw_data" in folder:
            readme_path.write_text("IMPORTANT: Never modify files in this folder!\nThis is your data vault - keep originals safe here.")
    
    # Add README for scripts folder
    scripts_readme = scripts_path / "README.txt"
    scripts_readme.write_text("Version-controlled scripts and code.\nUse git to track changes to your analysis code.")
    
create_project_structure("agrivoltaics_in_ri")
```
